```python
limit = 7000
emb = data[:limit]
from sklearn.manifold import TSNE
model = TSNE(n_components=2,random_state=1, perplexity=50)
edata = model.fit_transform(emb)
```

```python
label_names = {
    0:"0 Undefined",
    1:"1 Normal",
    2:"2 ESSV (PAC)",
    3:"3 Aberrated",
    4:"4 ESV (PVC)"
}
```

```python
plt.figure(figsize=(15,10))
selected_labels = labels2["label"][:limit]
X_embedded = edata
np.random.seed(0)
markers = ["+","o","<",">","*",".",",","p","3"]
colors = ["red","blue","green","yellow", "orange"]
for cat in np.unique(selected_labels):
    i_cat = np.where(selected_labels == cat)
    
    plt.scatter(X_embedded[i_cat][:,0], X_embedded[i_cat][:,1], 
                label=label_names[cat], 
                marker=markers[cat], 
                color=colors[cat]);
plt.legend();
```
